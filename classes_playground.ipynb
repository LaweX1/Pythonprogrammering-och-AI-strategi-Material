```python
class Person:
    def __init__(self, name, age, like_coding = "Nope!!"):
        self.name = name
        self.age = age
        self.like_coding = like_coding
    
    def presentation(self):
        print(f"Hello my name is {self.name} and I am {self.age}yrs old.\nGuess if I like coding? answer is: {self.like_coding}")

person_one = Person("Mark", 27)
person_one.like_coding = "Yes I love it!! :)"
person_one.presentation()
```

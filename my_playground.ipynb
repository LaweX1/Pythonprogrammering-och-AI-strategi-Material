```python
# - Dict -
my_dict = {"pears": 6, "bananas": 3, "apples": 10}

# Using get seems safer because if value does not exist we'll get 'None' value.
print(my_dict.get("pearszx"))
print(my_dict.get("pears"))

# Adding new Key / Value pair
my_dict["lemon"] = 9
# Updating value
my_dict["bananas"] = 7
# Checking all the keys
my_dict.keys()

# dict comprehension & items
# [(key, val) for key, val in my_dict.items()]
print("\n".join([f"{key.title()}: {val}" for key, val in my_dict.items()]))

# zip function
fruits = ["apples", "oranges", "bananas"]
amounts = (3, 6, 9)
new_fruit_dict = [(fruit, amount) for fruit, amount in zip(fruits, amounts)]
print(new_fruit_dict)
```
